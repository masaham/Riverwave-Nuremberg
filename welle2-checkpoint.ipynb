```python
import requests
import lxml.html as lh
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#url='https://www.hnd.bayern.de/pegel/regnitz/nuernberg-lederersteg-24225000/tabelle?methode=wasserstand&days=365'#Create a handle, page, to handle the contents of the website
url='https://www.gkd.bayern.de/de/fluesse/abfluss/kelheim/nuernberg-lederersteg-24225000/messwerte/tabelle?zr=woche&addhr=hr_hw&beginn=28.04.1990&ende=17.06.2020'

#url="https://www.gkd.bayern.de/de/fluesse/wasserstand/kelheim/nuernberg-lederersteg-24225000/messwerte/tabelle?zr=woche&addhr=hr_w_hw&beginn=29.04.1970&ende=28.04.2020"
#"https://www.gkd.bayern.de/de/fluesse/wasserstand/regnitz/huettendorf-24201501/messwerte/tabelle?zr=woche&addhr=hr_w_hw&beginn=01.01.1970&ende=28.04.2020"

page = requests.get(url)#Store the contents of the website under doc
doc = lh.fromstring(page.content)#Parse data that are stored between <tr>..</tr> of HTML
tr_elements = doc.xpath('//tr')
```

```python
#Check the length of the first 12 rows
[len(T) for T in tr_elements[:12]]
#doc.xpath('//tr')
tr_elements[1][1].text_content()
```

```python
tr_elements = doc.xpath('//tr')#Create empty list
col=[]
i=0#For each row, store each first element (header) and an empty list
for t in tr_elements[1]:
    i+=1
    name=t.text_content()
    print (i,name) #'%d:"%s"%'
    col.append((name,[]))
```

```python
#Since out first row is the header, data is stored on the second row onwards
for j in range(2,len(tr_elements)):
    #T is our j'th row
    T=tr_elements[j]
    
    #If row is not of size 2, the //tr data is not from our table 
    if len(T)!=4:
        break
    
    #i is the index of our column
    i=0
    
    #Iterate through each element of the row
    for t in T.iterchildren():
        data=t.text_content() 
        #Check if row is empty
        if i>0:
        #Convert any numerical value to integers
            try:
                data=int(data)
            except:
                pass
        #Append the data to the empty list of the i'th column
        col[i][1].append(data)
        #Increment i for the next column
        i+=1
```

```python
[len(C) for (title,C) in col]
```

```python
Dict={title:column for (title,column) in col}
df=pd.DataFrame(Dict)
```
